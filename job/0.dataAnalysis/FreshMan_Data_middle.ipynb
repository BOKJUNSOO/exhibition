```python
import pandas as pd
df = pd.read_csv(r"C:\Users\brian\Desktop\JUNSOO\exhibition\data\FreshMan.csv" , encoding='cp949')

## 초중고 의미있는 칼럼 추출

df_ele_col = ['지역'
               ,'적령 아동-합계', '유예 및 과령 아동-합계'
               ,'조기입학아동-합계'
               ]       #초등학교 데이터 열

df_MH_col = ['지역'
                 ,'당해연도 졸업자-합계'
                 ,'이전연도 졸업자-합계'
                 ,'검정고시 및 교육과정 이수자-합계'
                 ]      #중학교 고등학교는 데이터 열 동일

## 데이터 프레임 정의
df_ele_df = df[df_ele_col]
df_middle_df = df[df_MH_col]
df_high_df = df[df_MH_col]

# 2021 데이터
df_2021_ele = df_ele_df.iloc[0:603,:]
df_2021_middle = df_middle_df.iloc[603:992,:]
df_2021_high = df_high_df.iloc[992:1317,:]

# 2022 데이터
df_2022_ele = df_ele_df.iloc[1408:2010,:]
df_2022_middle = df_middle_df.iloc[2010:2405,:]
df_2022_high = df_high_df.iloc[2435:2763,:]

#2023 데이터
df_2023_ele1 = df_ele_df.iloc[2405:2435,:]
df_2023_ele2 = df_ele_df.iloc[2763:2780,:]
df_2023_ele3 = df_ele_df.iloc[2868:3418,:]      #2023초등학교 데이터가 순서대로 정렬되어 있지 않음
concat_2023_ele = pd.concat([df_2023_ele1,
                         df_2023_ele2,
                         df_2023_ele3])

df_2023_middle = df_middle_df.iloc[3418:3815,:]
df_2023_high = df_high_df.iloc[3841:4170,:]

#2024 데이터
df_2024_ele1 = df_ele_df.iloc[3815:3841,:]
df_2024_ele2 = df_ele_df.iloc[4170:4182,:]
df_2024_ele3 = df_ele_df.iloc[4272:4827,:]      #2024초등학교 데이터가 순서대로 정렬되어 있지 않음
concat_2024_ele = pd.concat([df_2024_ele1,
                         df_2024_ele2,
                         df_2024_ele3])


df_2024_middle = df_middle_df.iloc[4827:5215,:]
df_2024_high = df_high_df.iloc[5215:5538,:]
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'
```

```python
#중학교 통계량 - 당해연도 졸업자-합계 
 #2021
df_2021_middle_gradu = (df_2021_middle
                        .groupby(['지역'])['당해연도 졸업자-합계']
                        .sum())

df_2021_middle_gradu.plot(kind ='bar' , x = '지역', y = '당해연도 졸업자-합계'
             , figsize = (10,6) , color = 'skyblue')

plt.title('2021연도 중학교 졸업자-합계')
plt.xlabel('지역')
plt.ylabel('2021연도 중학교 졸업자-합계')
```
