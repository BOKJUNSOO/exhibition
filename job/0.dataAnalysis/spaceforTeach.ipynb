```python
import pandas as pd
df = pd.read_csv(r"C:\Users\brian\Desktop\JUNSOO\exhibition\data\SchoolTeacher.csv" , encoding='cp949')
```

```python
df['지역'] = df['지역'].fillna('서울특별시 기타')
df.info()
```

```python
## 교수학습 공간 일반교실
df_general = df['교수학습공간 일반교실']
(df_general
      .fillna(0 , inplace = True) # 없는 교실
      )
df_general = df[['지역', '교수학습공간 일반교실', '공시연도']]
#교수학습공간 일반교실 plot
df_2021_general = df_general[df_general['공시연도'] == 2021][['지역', '교수학습공간 일반교실']] 
df_2022_general = df_general[df_general['공시연도'] == 2022][['지역', '교수학습공간 일반교실']]
df_2023_general = df_general[df_general['공시연도'] == 2022][['지역', '교수학습공간 일반교실']]
df_2024_general = df_general[df_general['공시연도'] == 2022][['지역', '교수학습공간 일반교실']]

## 교수학습공간 교과교실
df_subject = df['교수학습공간 교과교실']
(df_subject
        .fillna(0 , inplace = True) # 없는 교실
        )
df_subject = df[['지역', '교수학습공간 교과교실', '공시연도']]
#교수학습공간 교과교실 plot
df_2021_subject = df_subject[df_subject['공시연도'] == 2021][['지역', '교수학습공간 교과교실']]
df_2022_subject = df_subject[df_subject['공시연도'] == 2022][['지역', '교수학습공간 교과교실']]
df_2023_subject = df_subject[df_subject['공시연도'] == 2023][['지역', '교수학습공간 교과교실']]
df_2024_subject = df_subject[df_subject['공시연도'] == 2024][['지역', '교수학습공간 교과교실']]

## 교수학습공간 특별교실
df_special = df['교수학습공간 특별교실']
(df_special
        .fillna(0 , inplace = True)) #없는 교실
df_special = df[['지역', '교수학습공간 특별교실', '공시연도']]

# 교수학습공간 특별교실 plot
df_2021_special = df_special[df_special['공시연도'] == 2021][['지역', '교수학습공간 특별교실']]
df_2022_special = df_special[df_special['공시연도'] == 2022][['지역', '교수학습공간 특별교실']]
df_2023_special = df_special[df_special['공시연도'] == 2023][['지역', '교수학습공간 특별교실']]
df_2024_special = df_special[df_special['공시연도'] == 2024][['지역', '교수학습공간 특별교실']]

## 교수학습공간 수준별 교실
df_level = df['교수학습공간 수준별교실']
(df_level
        .fillna(0 , inplace = True)) #없는 교실
df_level = df[['지역', '교수학습공간 수준별교실', '공시연도']]

#교수학습공간 수준별교실 plot
df_2021_level = df_level[df_level['공시연도'] == 2021][['지역', '교수학습공간 수준별교실']]
df_2022_level = df_level[df_level['공시연도'] == 2022][['지역', '교수학습공간 수준별교실']]
df_2023_level = df_level[df_level['공시연도'] == 2023][['지역', '교수학습공간 수준별교실']]
df_2024_level = df_level[df_level['공시연도'] == 2024][['지역', '교수학습공간 수준별교실']]

print(df_2024_general['교수학습공간 일반교실'].sum()
      ,df_2024_level['교수학습공간 수준별교실'].sum()
      ,df_2024_special['교수학습공간 특별교실'].sum()
      ,df_2024_subject['교수학습공간 교과교실'].sum()
      ,sep = '\n')
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'

#2024교수학습공간 일반교실
df_2024_general_g = (df_2024_general
               .groupby(['지역'])['교수학습공간 일반교실']
               .sum())

df_2024_general_g.plot(kind ='bar' , x = '지역', y = '교수학습공간 일반교실'
             , figsize = (10,6) , color = 'skyblue')

plt.title('2024 교수학습공간 일반교실')
plt.xlabel('지역')
plt.ylabel('2024 교수학습공간 일반교실')



```

```python
#교수학습공간 교과교실
df_2024_subject_g = (df_2024_subject
               .groupby(['지역'])['교수학습공간 교과교실']
               .sum())

df_2024_subject_g.plot(kind ='bar' , x = '지역', y = '교수학습공간 교과교실'
             , figsize = (10,6) , color = 'pink')

plt.title('2024 교수학습공간 교과교실')
plt.xlabel('지역')
plt.ylabel('2024 교수학습공간 교과교실')
```
