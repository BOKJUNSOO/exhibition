```python
# 학교별 입학생 현황
import pandas as pd
df = pd.read_csv(r"C:\Users\brian\Desktop\JUNSOO\exhibition\data\FreshMan.csv" , encoding='cp949')
#df.info() #초중고별 제공하는 데이터가 다름
#'지역' 칼럼에서 "서울특별시" 라는 단어 삭제 Ctrl+H

#df_1 = 초등학교
#df_2 = 중학교
#df_3 = 고등학교

#데이터 프레임 정의
#초등학교
col1 = ['지역'
        ,'적령 아동-합계'
        ,'유예 및 과령 아동-합계'
        ,'조기입학아동-합계'
        ,'합계']
df_1 = df[col1]
df_elem = df_1.iloc[0:605,:]

#605 ~ 993 중학교
col2 = ['지역'
           ,'당해연도 졸업자-합계'
           ,'이전연도 졸업자-합계'
           ,'검정고시 및 교육과정 이수자-합계'
           , '합계(2020년 이전 : A+B)']

col2_find_na = ['당해연도 졸업자-합계'          #지역 데이터 제외하고 결측치 찾기
           ,'이전연도 졸업자-합계'
           ,'검정고시 및 교육과정 이수자-합계'
           , '합계(2020년 이전 : A+B)']
df_2 = df[col2]
df_middle = df_2.iloc[605:994,:]

#994~ 1318 고등학교 (중학교와 기록된 데이터의 columns 동일함)
df_high = df_2.iloc[994:1319,:]


#초등학교 데이터 결측값 처리
print(df_elem
      .isnull()
      .sum()) #603, #604 row

print(df_elem[['적령 아동-합계','유예 및 과령 아동-합계'
        , '조기입학아동-합계']]
        .mean())

df_elem['적령 아동-합계'] = df_elem['적령 아동-합계'].fillna(105.17578772802653)
df_elem['유예 및 과령 아동-합계'] = df_elem['유예 및 과령 아동-합계'].fillna(0.276949)
df_elem['조기입학아동-합계'] = df_elem['조기입학아동-합계'].fillna(0.169154)

df_elem.isnull().sum()
```

```python
#중학교 고등학교 데이터 결측값 처리 ==> 고등학교 데이터에 결측값 존재
df_high[col2_find_na].mean()
df_high['당해연도 졸업자-합계'] = df_high['당해연도 졸업자-합계'].fillna(206.904025)
df_high['검정고시 및 교육과정 이수자-합계'] = df_high['검정고시 및 교육과정 이수자-합계'].fillna(1.566563)
df_high['합계(2020년 이전 : A+B)'] = df_high['합계(2020년 이전 : A+B)'].fillna(209.894737)

df_high.isnull().sum()
      
```

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import matplotlib.font_manager as fm

plt.rcParams['font.family'] = 'Malgun Gothic'

#초등학교 통계량
 #적령 아동-합계
df_elem_fit = (df_elem
               .groupby(['지역'])['적령 아동-합계']
               .sum())

df_elem_fit.plot(kind ='bar' , x = '지역', y = '적령 아동-합계'
             , figsize = (10,6) , color = 'blue')
 

plt.title('지역별 적령 아동수')
plt.xlabel('지역')
plt.ylabel('적령아동수')
```

```python
 #유예 및 과령 아동-합계
df_elem_over = (df_elem
               .groupby(['지역'])['유예 및 과령 아동-합계']
               .sum())

df_elem_over.plot(kind ='bar' , x = '지역', y = '유예 및 과령 아동-합계'
             , figsize = (10,6) , color = 'blue')

plt.title('유예 및 과령 아동-합계')
plt.xlabel('지역')
plt.ylabel('유예 및 과령 아동-합계')
```

```python
 #조기입학아동-합계
df_elem_early = (df_elem
               .groupby(['지역'])['조기입학아동-합계']
               .sum())

df_elem_early.plot(kind ='bar' , x = '지역', y = '조기입학아동-합계'
             , figsize = (10,6) , color = 'blue')

plt.title('조기입학아동-합계')
plt.xlabel('지역')
plt.ylabel('조기입학아동-합계')
```

```python
#중학교 통계량
 #당해연도 졸업자-합계 #칼럼이름은 당해이지만 2021~2024 년도까지의 졸업자 합계
                     
df_middle_gradu = (df_middle
               .groupby(['지역'])['당해연도 졸업자-합계']
               .sum())

df_middle_gradu.plot(kind ='bar' , x = '지역', y = '당해연도 졸업자-합계'
             , figsize = (10,6) , color = 'blue')

plt.title('당해연도 졸업자-합계')
plt.xlabel('지역')
plt.ylabel('당해연도 졸업자-합계')
```
