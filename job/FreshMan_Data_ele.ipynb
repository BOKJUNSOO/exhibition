```python
import pandas as pd
df = pd.read_csv(r"C:\Users\brian\Desktop\JUNSOO\exhibition\data\FreshMan.csv" , encoding='cp949')
```

```python
## 초중고 의미있는 칼럼 추출

df_ele_col = ['지역'
               ,'적령 아동-합계', '유예 및 과령 아동-합계'
               ,'조기입학아동-합계'
               ]       #초등학교 데이터 열

df_MH_col = ['지역'
                 ,'당해연도 졸업자-합계'
                 ,'이전연도 졸업자-합계'
                 ,'검정고시 및 교육과정 이수자-합계'
                 ]      #중학교 고등학교는 데이터 열 동일

## 데이터 프레임 정의
df_ele_df = df[df_ele_col]
df_middle_df = df[df_MH_col]
df_high_df = df[df_MH_col]

# 2021 데이터
df_2021_ele = df_ele_df.iloc[0:603,:]
df_2021_middle = df_middle_df.iloc[605:993,:]
df_2021_high = df_high_df.iloc[994:1317,:]

# 2022 데이터
df_2022_ele = df_ele_df.iloc[1410:2010,:]
df_2022_middle = df_middle_df.iloc[2012:2405,:]
df_2022_high = df_high_df.iloc[2437:2763,:]

#2023 데이터
df_2023_ele1 = df_ele_df.iloc[2407:2435,:]
df_2023_ele2 = df_ele_df.iloc[2765:2780,:]
df_2023_ele3 = df_ele_df.iloc[2870:3418,:]      #2023초등학교 데이터가 순서대로 정렬되어 있지 않음
concat_2023_ele = pd.concat([df_2023_ele1,
                         df_2023_ele2,
                         df_2023_ele3])

df_2023_middle = df_middle_df.iloc[3420:3815,:]
df_2023_high = df_high_df.iloc[3843:4170,:]

#2024 데이터
df_2024_ele1 = df_ele_df.iloc[3817:3841,:]
df_2024_ele2 = df_ele_df.iloc[4172:4182,:]
df_2024_ele3 = df_ele_df.iloc[4274:4827,:]      #2024초등학교 데이터가 순서대로 정렬되어 있지 않음
concat_2024_ele = pd.concat([df_2024_ele1,
                         df_2024_ele2,
                         df_2024_ele3])


df_2024_middle = df_middle_df.iloc[4829:5215,:]
df_2024_high = df_high_df.iloc[5217:5538,:]

## 결측값 처리
#초등학교
concat_2024_ele['적령 아동-합계'] = concat_2024_ele['적령 아동-합계'].fillna(105)
concat_2024_ele['유예 및 과령 아동-합계'] = concat_2024_ele['유예 및 과령 아동-합계'].fillna(0)
concat_2024_ele['조기입학아동-합계'] = concat_2024_ele['조기입학아동-합계'].fillna(0)

#중학교
df_2022_middle['당해연도 졸업자-합계'] = df_2022_middle['당해연도 졸업자-합계'].fillna(207)
df_2022_middle['이전연도 졸업자-합계'] = df_2022_middle['이전연도 졸업자-합계'].fillna(1)
df_2022_middle['검정고시 및 교육과정 이수자-합계'] = df_2022_middle['검정고시 및 교육과정 이수자-합계'].fillna(2)

df_2023_high['당해연도 졸업자-합계'] = df_2023_high['당해연도 졸업자-합계'].fillna(207)
df_2023_high['이전연도 졸업자-합계'] = df_2023_high['이전연도 졸업자-합계'].fillna(1)
df_2023_high['검정고시 및 교육과정 이수자-합계'] = df_2023_high['검정고시 및 교육과정 이수자-합계'].fillna(2)



#고등학교
df_2022_high['당해연도 졸업자-합계'] = df_2022_high['당해연도 졸업자-합계'].fillna(207)
df_2022_high['이전연도 졸업자-합계'] = df_2022_high['이전연도 졸업자-합계'].fillna(1)
df_2022_high['검정고시 및 교육과정 이수자-합계'] = df_2022_high['검정고시 및 교육과정 이수자-합계'].fillna(2)

df_2023_high['당해연도 졸업자-합계'] = df_2023_high['당해연도 졸업자-합계'].fillna(207)
df_2023_high['이전연도 졸업자-합계'] = df_2023_high['이전연도 졸업자-합계'].fillna(1)
df_2023_high['검정고시 및 교육과정 이수자-합계'] = df_2023_high['검정고시 및 교육과정 이수자-합계'].fillna(2)

df_2024_high['당해연도 졸업자-합계'] = df_2024_high['당해연도 졸업자-합계'].fillna(207)
df_2024_high['이전연도 졸업자-합계'] = df_2024_high['이전연도 졸업자-합계'].fillna(1)
df_2024_high['검정고시 및 교육과정 이수자-합계'] = df_2024_high['검정고시 및 교육과정 이수자-합계'].fillna(2)
```
